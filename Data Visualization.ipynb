```python
### Total units of product sold

# Frequency Table:

# First, let us understand what a frequency distribution table looks like. 
# In Statistics, frequency is the number of occurrences of each item. 
# It has two columns, the category itself and the corresponding frequency. 
# By organizing the data in this manner, we can see Colorbuds is highest selling product. 

df.groupby('Product_Name')['Units'].sum()
```

```python
df.groupby('Product_Name')['Units'].sum().sort_values(ascending=False)
```

```python
# **Simple Bar Chart:** 
    
# If you want to represent the data based using one or two variables, then the simple bar diagram can be used. 
# A simple bar chart visually represents data using rectangular bars of equal width.

# Group by 'Product_Name', sum the 'Units', and reset index to convert it back to a DataFrame
units_by_product_df= df.groupby('Product_Name')['Units'].sum().reset_index()

# Sort the DataFrame by 'Total_Units_Sold' in descending order
units_by_product_df_sorted=units_by_product_df.sort_values(by='Units',ascending=False)


# Plotting the bar chart
plt.figure(figsize=(16, 6))  # Adjust the size as needed
plt.bar(units_by_product_df_sorted['Product_Name'], units_by_product_df_sorted['Units'], color='green')

# Adding title and labels
plt.title('Total Units Sold Per Product')
plt.xlabel('Product Name')
plt.ylabel('Total Units Sold')
```

```python
# # **Pie Chart:**
    
#  A pie diagram/chart is used when the requirement of the situation is to know the 
# relationship between the whole of a thing and its parts
#  While making comparisons, pie diagrams should be used on a percentage basis and not on an absolute basis. 
#  Pie charts are useful to understand the share of each category to the total.
#  In Statistics, this is known as relative frequency. 
#All relative frequencies should add up to 100%. Pie charts are useful to understand the share of each category to the total.
```

```python
# Group by 'Product_Name', sum the 'Units', and reset index to convert it back to a DataFrame
units_by_product_df = df.groupby('Product_Name')['Units'].sum().reset_index()
# Plotting the pie chart
plt.figure(figsize=(10, 8))  # Adjust the size as needed
plt.pie(units_by_product_df['Units'], labels=units_by_product_df['Product_Name'], autopct='%1.1f%%')
# autopct='%1.1f%%': This is an optional argument that automatically adds a percentage to each slice of the pie chart, formatted to one decimal place. The '%1.1f%%' is a Python string format specification; it means that the number will be formatted with one digit after the decimal point and followed by a percent sign.

# Adding title
plt.title('Proportion of Total Units Sold by Product')

# Show the plot
plt.show()
```

```python
# ### Total units of product sold by each store
# **Contingency table:**

# Contingency tables (also called crosstabs or two-way tables) are used in statistics to 
# summarize the relationship between severalcategorical variables.
# It helps us understand how these variables are related and provides a clear overview of the distribution of categories 
# for each variable.


# Creating a contingency table
# The crosstab function is used to compute a simple cross-tabulation of two (or more) factors. 
#It's essentially used to create a contingency table, which shows the frequency with which certain groups of data appear.
pd.crosstab(df['Product_Name'], df['Store_ID'])
```

```python
# ### Total units of product sold by each store

# **Multiple Bar Diagram** :

# Multiple bar diagram is used for comparing two or more sets of data. 
# Bars are constructed side by side to represent the set of values for comparison.


grouped_data = df.groupby(['Product_Name', 'Store_ID'])['Units'].sum().reset_index()

plt.figure(figsize=(15, 5))
# Creating the bar plot
sns.barplot(x='Product_Name', y='Units', hue='Store_ID', data=grouped_data)

# Adding labels and title
plt.xlabel('Toy')
plt.ylabel('Total Units Sold')
plt.title('Total Units of Each Toy Sold by Each Store')

# Displaying the plot
plt.show()


#We can clearly see, Store 1 has the highest sales in all product sold. 
```

```python
# ### Distribution of product category sold by each store

# **Sub-divided Bar Diagram**

# If various components of a variable are to be represented in a single diagram, 
#then subdivided bar diagrams are made in this situation. 
# A subdivided bar diagram drawn based on the percentage of the total is known as a percentage bar diagram.

grouped_data = df.groupby(['Store_ID', 'Product_Category'])['Units'].sum().unstack()

normalized_data = grouped_data.div(grouped_data.sum(axis=1), axis=0) * 100

ax = normalized_data.plot(kind='bar', stacked=True, figsize=(12, 8), width=0.4)

# Adding labels and title
plt.xlabel('Store ID')
plt.ylabel('Percentage of Total Units Sold (%)')
plt.title('Distribution of Product Categories Sold by Each Store (Percentage)')
plt.xticks(rotation=360)
        
# Adding the legend
plt.legend(title='Product Category', bbox_to_anchor=(1.02, 1), loc='upper left')

# Displaying the plot
plt.tight_layout()  # Adjusts plot to ensure everything fits without overlapping
plt.show()
```

```python
### Quarterly revenue generated between all the stores

#**Line Chart**:

#A line Chart is used to show information change over time. 

# Ensure the Date column is in datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Create the 'Year_Quarter' column
df['Year_Quarter'] = df['Date'].dt.to_period('Q').astype(str)

# Display the DataFrame to verify the new column
df.tail()
```

```python
df.groupby(['Year_Quarter', 'Store_ID'])['Revenue'].sum().unstack()
```

```python
quarterly_revenue_by_store = df.groupby(['Year_Quarter', 'Store_ID'])['Revenue'].sum().unstack()

# Setting up the plot
plt.figure(figsize=(10, 5))

# Plotting a line for each store
for store_id in quarterly_revenue_by_store.columns:
    plt.plot(quarterly_revenue_by_store.index, quarterly_revenue_by_store[store_id], marker='o', label=f'{store_id}')

# Adding title and labels
plt.title('Quarterly Revenue Generated by Each Store')
plt.xlabel('Year and Quarter')
plt.ylabel('Total Revenue')

# Adding grid lines and legend
plt.grid(True)
plt.legend(title='Store ID')

# Display the plot
plt.tight_layout()  # Adjust layout to fit everything neatly
plt.show()
```

```python
# ### Distribution of sum of revenue generated per day

# **Histogram:**

# A histogram is a chart showing the frequency of occurrence of each value of the variable being analysed.
# In a histogram, data are plotted as a series of rectangles. 
# Class intervals are shown on the X-axis and the frequencies are on the Y-axis. 
# The height of each rectangle represents the frequency of the class interval. 
# Each rectangle is formed with the other to give a continuous picture.


# Group by 'Date' and sum the 'Revenue'
daily_revenue = df.groupby(df['Date']).sum()['Revenue']

# Now, plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(daily_revenue, bins=20, alpha=0.7, color='blue', edgecolor='black')

# Adding titles and labels
plt.title('Distribution of Sum of Revenue Generated Per Day')
plt.xlabel('Sum of Revenue')
plt.ylabel('Frequency of Sum per Day')

# Adding grid lines for better readability
plt.grid(axis='y', alpha=0.75)

# Show the plot
plt.xlim(left=0)
plt.tight_layout()
plt.show()
```

```python
# ### Check daily distribution of revenue generated

# **Boxplot**:

# A box plot is a method for graphically depicting groups of numerical data through their quartiles. 
# Boxplots are a standardized way of displaying the distribution of data based on a 
# five-number summary (minimum, first quartile (Q1), median, third quartile (Q3), and maximum). 
# The Box-plot method is used to identify the outliers in the data.


# Group by 'Date' and sum the 'Revenue'
daily_revenue = df.groupby('Date')['Revenue'].sum()

# Create a box plot for the daily revenue
plt.figure(figsize=(10, 6))
plt.boxplot(daily_revenue)
plt.title('Box Plot of Sum of Revenue Generated Per Day')
plt.xlabel('Day')
plt.ylabel('Sum of Revenue')

# Show the plot
plt.show()

# Min --> Q1- 1.5*IQR
# Max --> Q3+ 1.5*IQR
```

```python
# ### Find the relationship between no. of units placed on a daily basis and the revenue generated

# **Scatter plot**:

# A scatter plot (aka scatter chart, scatter graph) uses dots to represent values for two different numeric variables.
# The dots in a scatter plot not only report the values of individual data points 
# but also patterns when the data are taken as a whole.


# Group by 'Date' and calculate the sum of 'Units' and 'Revenue'
daily_data = df.groupby('Date').agg({'Units': 'sum', 'Revenue': 'sum'})

# Now create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(daily_data['Units'], daily_data['Revenue'], alpha=0.5)

# Adding titles and labels
plt.title('Daily Units vs. Revenue Generated')
plt.xlabel('Number of Units Placed')
plt.ylabel('Revenue Generated')

# Show the plot
plt.show()
```

```python
# ### Identify highest selling products that generated maximum revenue

# **Pareto Chart**:
    
# A Pareto diagram is a special type of bar chart, where the categories are shown in descending order of frequency. 
# The Pareto Principle, also known as the 80/20 Rule, states that 80% of the outcomes come from 20% of the actions.

# Calculate revenue per product
revenue_per_product = df.groupby('Product_Name')['Revenue'].sum().sort_values(ascending=False)

# Calculate the cumulative percentage
cumulative_percentage = revenue_per_product.cumsum() / revenue_per_product.sum() * 100

# Create a new figure and a twin Axes sharing the xaxis
fig, ax = plt.subplots(figsize=(14, 7))
ax.bar(revenue_per_product.index, revenue_per_product, color='skyblue')

# Make the y-axis label, ticks and tick labels match the line plot
ax2 = ax.twinx()
ax2.plot(revenue_per_product.index, cumulative_percentage, color='orange', marker='o')
ax.set_title('Pareto Chart of Highest Selling Products')
ax.set_xlabel('Product Name')
ax.set_ylabel('Revenue')
ax2.set_ylabel('Cumulative Percentage (%)')
ax2.set_ylim(0, 105)

# Adding a horizontal line at 80% on the cumulative percentage to indicate the "vital few"
ax2.axhline(y=80, color='red', linestyle='--')

# Show the plot
plt.show()
```
